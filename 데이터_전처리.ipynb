```python
# Answer1부터 Answer5까지 병합
answers = finetuning[['Answers1', 'Answers2', 'Answers3', 'Answers4', 'Answers5']].values.flatten()

# summary1부터 summary5까지 병합
summaries = finetuning[['summary1', 'summary2', 'summary3', 'summary4', 'summary5']].values.flatten()

# 새로운 데이터 프레임 생성
merged_df = pd.DataFrame({
    'Answer': answers,
    'summary': summaries
})
```

```python
from rouge import Rouge

def evaluate_summary(original, summary):
    rouge = Rouge()
    scores = rouge.get_scores(summary, original)
    return scores[0]['rouge-1']['f']  # Example using ROUGE-1 F1 score

# CSV 파일 로드 (이미 요약본이 추가된 데이터셋)
df = pd.read_csv('/content/filtered_data.csv')

df['rouge_score'] = df.apply(lambda x: evaluate_summary(x['Answer'], x['summary']), axis=1)

threshold = 0.7  # 적절한 임계값 설정
high_quality_summaries = df[df['rouge_score'] > threshold]

high_quality_summaries.to_csv('high_quality_summaries.csv', index=False)
```
